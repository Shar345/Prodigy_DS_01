```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('taxis')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
#Visualize the Continuous variable 
plt.hist(x = data['fare'], 
         color='blue', 
         bins = [0,5,10,15,20,25,30,35,40,50],
         histtype='stepfilled', edgecolor="black", 
         alpha=0.5)
plt.xlabel("fare")
plt.ylabel("No of Passengers")
plt.title("Distribution of fare")
```

```python
#Visualize the categorical variable 
sns.countplot(x = data['payment'])
plt.xlabel("Payment")
plt.ylabel("No of passengers")
plt.title("Distribution of Payment type")
```
